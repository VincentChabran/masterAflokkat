```python
import tensorflow as tf
import skimage as ski
from preprocessing import load_processed_dataset
import matplotlib.pyplot as plt
import random
import numpy as np
```

```python
x, y = load_processed_dataset()
```

```python
x = np.asarray(x)
y = np.asarray(y)
```

```python
model = tf.keras.Sequential([
    tf.keras.layers.Conv2D(15, 3, activation="relu", input_shape=(50,50,1)),
    # tf.keras.layers.Conv2D(10, 3, activation="relu"),
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(100, activation="relu"),
    tf.keras.layers.Dense(50, activation="relu"),
    tf.keras.layers.Dense(1, activation="sigmoid")
])
```

```python
model.compile(optimizer="adam",
              loss="binary_crossentropy",
              metrics=["accuracy"])
```

```python
model.fit(x,y, epochs=15)
```

```python
from preprocessing import process_image

im = process_image(ski.io.imread("random images/chien5.jpeg"))
plt.imshow(im, cmap="gray")
plt.show()

model.predict(np.asarray([im]))
```
