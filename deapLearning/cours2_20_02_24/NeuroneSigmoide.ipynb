```python
from numpy import exp
import numpy


def sigmoide(z):
   return 1/(1 + exp(-z))


class NueroneSigomoide:
   def __init__(self, poids, biais) -> None:
      self.poids = poids 
      self.biais = biais

   def forward(self, entrées): 
      z = 0

      # Deuxiéme syntaxe a la place de la boucle for 
      # z = numpy.dot(entrées, self.poids)
      for i in range(len(self.poids)): 
         z += self.poids[i] * entrées[i]
      
      z += self.biais
      return sigmoide(z)
```

```python
poids = 54

neurone_nand = NueroneSigomoide([-poids, -poids], 100)

# print(neurone_nand.forward([-10, 10, 10]))

print(neurone_nand.forward([1,1]))
print(neurone_nand.forward([1,0]))
print(neurone_nand.forward([0,1]))
print(neurone_nand.forward([0,0]))
```
