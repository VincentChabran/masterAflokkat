**Question 3 :** Créez une chaîne de Markov à partir des résumés. Chaque sommet de la chaîne sera un mot, et une arête de poids _n_ relie le mot A au mot B si et seulement si le mot B suit le mot A _n_ fois au total dans les résumés.

```python
mon_dictionnaire = {}

for resume in resumes:
   mots = resume.split(" ")
   
   for mot in mots:
      for i, mot2 in enumerate(mots): 
         if mot == mot2 and i + 1 < len(mots):
            if mot in mon_dictionnaire:
               mon_dictionnaire[mot].append(mots[i+1])
            else:
               mon_dictionnaire[mot] = []
               mon_dictionnaire[mot].append(mots[i+1])
               

len(mon_dictionnaire["jour"])
```
