```python
from pyspark import SparkContext

!hadoop fs -put /vagrant/tp/3/friends.txt
!hadoop fs -ls friends.txt

friends = sc.textFile('friends.txt')
friends.collect()

user_friends = friends.map(lambda line: line.split("=>"))
user_friends.collect()

myList = user_friends.map(lambda x: (x[0], x[1].split(",")))
myList.collect()

user_pairs = myList.flatMap(lambda x: [((x[0], friend), set(x[1])) for friend in x[1]])
user_pairs.collect()

sorted_user_pairs = user_pairs.map(lambda x: (tuple(sorted(x[0])), x[1]))
sorted_user_pairs.collect()

common_friends = sorted_user_pairs.reduceByKey(lambda x, y: x.intersection(y))
common_friends.collect()
```
